```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from IPython.display import display, Math
```

```python
def u(t):
    return np.where(t >= 0, 1, 0)
```

```python
display(Math(r"\textbf{Questão 1}"))
display(Math(r"x(t)=(t^2-1)\,[u(t-1)-u(t-2)]"))
```

```python
def x1(t):
    return (t**2 - 1) * (u(t-1) - u(t-2))
```

```python
display(Math(r"\textbf{(a) Plote o sinal } x(t) \textbf{ para } 0 \le t \le 3"))

t = np.linspace(0, 3, 1000)
xt = x1(t)

plt.figure(figsize=(10,4))
plt.plot(t, xt, linewidth=2)
plt.axhline(0, color="black", linewidth=0.8)
plt.grid(True)
plt.title("Questão 1(a) - Sinal x(t)")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.show()
```

```python
display(Math(r"\textbf{(b) Energia analítica do sinal}"))
display(Math(r"E=\int_{-\infty}^{\infty}|x(t)|^2\,dt"))
display(Math(r"x(t)\neq 0 \text{ apenas em } 1\le t<2"))
display(Math(r"E=\int_{1}^{2}(t^2-1)^2dt"))
display(Math(r"(t^2-1)^2=t^4-2t^2+1"))
display(Math(r"E=\int_{1}^{2}(t^4-2t^2+1)\,dt"))
display(Math(r"E=\left[\frac{t^5}{5}-\frac{2t^3}{3}+t\right]_{1}^{2}"))
display(Math(r"E=\frac{38}{15}\approx 2.5333"))

E_analitica = 38/15
```

```python
display(Math(r"\textbf{(c) Energia numérica do sinal}"))
display(Math(r"E_{num}\approx\sum |x(t)|^2\Delta t"))

dt = t[1] - t[0]
E_numerica = np.sum(np.abs(xt)**2) * dt

display(Math(rf"E_{{analítico}}=\frac{{38}}{{15}} \approx {E_analitica:.6f}"))
display(Math(rf"E_{{numérico}}\approx {E_numerica:.6f}"))
display(Math(rf"\text{{Erro absoluto}} \approx {abs(E_analitica - E_numerica):.6e}"))
```

```python
display(Math(r"\textbf{Questão 2}"))
display(Math(r"\text{O sinal é definido conforme a Figura 1.}"))

display(Math(r"x(t)=\begin{cases}"
             r"0, & t<-1.5 \\"
             r"2, & -1.5\le t < 0 \\"
             r"2e^{-t/2}, & 0\le t < 3 \\"
             r"0, & t\ge 3"
             r"\end{cases}"))

display(Math(r"x(t)=2[u(t+1.5)-u(t)] + 2e^{-t/2}[u(t)-u(t-3)]"))
```

```python
def x2(t):
    return 2*(u(t+1.5) - u(t)) + 2*np.exp(-t/2)*(u(t) - u(t-3))
```

```python
display(Math(r"\textbf{(b) Plotar o sinal } x(t)"))

t = np.linspace(-4, 6, 2000)
xt2 = x2(t)

plt.figure(figsize=(10,4))
plt.plot(t, xt2, linewidth=2)
plt.axhline(0, color="black", linewidth=0.8)
plt.grid(True)
plt.title("Questão 2(b) - Sinal x(t)")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.show()
```

```python
def plot_signal(t, y, title):
    plt.figure(figsize=(10,4))
    plt.plot(t, y, linewidth=2)
    plt.axhline(0, color="black", linewidth=0.8)
    plt.grid(True)
    plt.title(title)
    plt.xlabel("t")
    plt.ylabel("Amplitude")
    plt.show()
```

```python
display(Math(r"\textbf{(c-i) } x(-t)"))
display(Math(r"\text{Inversão no tempo (espelhamento em torno do eixo vertical).}"))

plot_signal(t, x2(-t), "Questão 2(c-i) - x(-t)")
```

```python
display(Math(r"\textbf{(c-ii) } x(t+4)"))
display(Math(r"\text{Deslocamento de 4 unidades para a esquerda.}"))

plot_signal(t, x2(t+4), "Questão 2(c-ii) - x(t+4)")
```

```python
display(Math(r"\textbf{(c-iii) } x(t-2)"))
display(Math(r"\text{Deslocamento de 2 unidades para a direita.}"))

plot_signal(t, x2(t-2), "Questão 2(c-iii) - x(t-2)")
```

```python
display(Math(r"\textbf{(c-iv) } x(2t)"))
display(Math(r"\text{Compressão no tempo por fator 2 (sinal mais rápido).}"))

plot_signal(t, x2(2*t), "Questão 2(c-iv) - x(2t)")
```

```python
display(Math(r"\textbf{(c-v) } x(t/2)"))
display(Math(r"\text{Expansão no tempo por fator 2 (sinal mais lento).}"))

plot_signal(t, x2(t/2), "Questão 2(c-v) - x(t/2)")
```

```python
display(Math(r"\textbf{Questão 4}"))
display(Math(r"h(t)=e^{-2t}u(t)"))
```

```python
def h(t):
    return np.exp(-2*t) * u(t)

def x4(t):
    return u(t) - u(t-2)
```

```python
display(Math(r"\textbf{(a) Plotar } h(t) \textbf{ para } 0\le t \le 3"))

t = np.linspace(0, 3, 1000)

plt.figure(figsize=(10,4))
plt.plot(t, h(t), linewidth=2)
plt.grid(True)
plt.title("Questão 4(a) - h(t) = e^(-2t)u(t)")
plt.xlabel("t")
plt.ylabel("h(t)")
plt.show()
```

```python
display(Math(r"\textbf{(b) Causalidade e estabilidade}"))

display(Math(r"\text{Causalidade: } h(t)=0 \text{ para } t<0 \Rightarrow \text{sistema causal}"))

display(Math(r"\text{Estabilidade: } \int_{-\infty}^{\infty}|h(t)|dt<\infty"))

display(Math(r"\int_{0}^{\infty} e^{-2t}dt=\frac{1}{2}<\infty \Rightarrow \text{sistema estável}"))
```

```python
display(Math(r"\textbf{(c) Convolução analítica}"))
display(Math(r"x(t)=u(t)-u(t-2)"))

display(Math(r"y(t)=x(t)*h(t)=\int_{-\infty}^{\infty}x(\tau)h(t-\tau)d\tau"))

display(Math(r"y(t)=\begin{cases}"
             r"0, & t<0 \\"
             r"\frac{1}{2}(1-e^{-2t}), & 0\le t<2 \\"
             r"\frac{1}{2}(e^{4}-1)e^{-2t}, & t\ge 2"
             r"\end{cases}"))
```

```python
def y_analitico(t):
    y = np.zeros_like(t)

    cond1 = (t >= 0) & (t < 2)
    cond2 = (t >= 2)

    y[cond1] = 0.5*(1 - np.exp(-2*t[cond1]))
    y[cond2] = 0.5*(np.exp(4) - 1)*np.exp(-2*t[cond2])

    return y
```

```python
display(Math(r"\textbf{(d) Convolução numérica e comparação com a analítica}"))

dt = 0.001
t = np.arange(-1, 6, dt)

xt = x4(t)
ht = h(t)

y_num = np.convolve(xt, ht) * dt
t_conv = np.arange(len(y_num)) * dt + (t[0] + t[0])

y_an = y_analitico(t_conv)

plt.figure(figsize=(12,5))
plt.plot(t_conv, y_num, label="Numérico", linewidth=2)
plt.plot(t_conv, y_an, "--", label="Analítico", linewidth=2)
plt.grid(True)
plt.title("Questão 4(d) - Comparação Numérico vs Analítico")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.legend()
plt.xlim(-1, 6)
plt.show()
```

```python
erro = np.mean(np.abs(y_num - y_an))
display(Math(rf"\text{{Erro médio absoluto}} \approx {erro:.6e}"))
```

```python
display(Math(r"\textbf{(e) Repetir para } h(t)=e^t u(t)"))
display(Math(r"h(t)=e^t u(t)"))
display(Math(r"\int_0^\infty e^t dt=\infty \Rightarrow \text{sistema instável}"))
```

```python
def h2(t):
    return np.exp(t) * u(t)
```

```python
t_plot = np.linspace(0, 3, 1000)

plt.figure(figsize=(10,4))
plt.plot(t_plot, h2(t_plot), linewidth=2)
plt.grid(True)
plt.title("Questão 4(e-a) - h(t) = e^t u(t)")
plt.xlabel("t")
plt.ylabel("h(t)")
plt.show()
```

```python
display(Math(r"y(t)=\begin{cases}"
             r"0, & t<0 \\"
             r"e^t-1, & 0\le t<2 \\"
             r"e^t(1-e^{-2}), & t\ge 2"
             r"\end{cases}"))
```

```python
def y2_analitico(t):
    y = np.zeros_like(t)

    cond1 = (t >= 0) & (t < 2)
    cond2 = (t >= 2)

    y[cond1] = np.exp(t[cond1]) - 1
    y[cond2] = np.exp(t[cond2])*(1 - np.exp(-2))

    return y
```

```python
dt = 0.001
t = np.arange(-1, 6, dt)

xt = x4(t)
ht2 = h2(t)

y_num2 = np.convolve(xt, ht2) * dt
t_conv2 = np.arange(len(y_num2)) * dt + (t[0] + t[0])

y2_an = y2_analitico(t_conv2)

plt.figure(figsize=(12,5))
plt.plot(t_conv2, y_num2, label="Numérico", linewidth=2)
plt.plot(t_conv2, y2_an, "--", label="Analítico", linewidth=2)
plt.grid(True)
plt.title("Questão 4(e-d) - Comparação Numérico vs Analítico")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.legend()
plt.xlim(-1, 6)
plt.show()
```

```python
erro2 = np.mean(np.abs(y_num2 - y2_an))
display(Math(rf"\text{{Erro médio absoluto}} \approx {erro2:.6e}"))
```

```python
display(Math(r"\textbf{(a) Solução analítica final}"))
display(Math(r"v_o(t)=\frac{1}{3}e^{-5t}+\frac{44}{3}e^{-2t}-9e^{-3t}"))
```

```python
t = np.linspace(0, 2, 1000)

vo_manual = (1/3)*np.exp(-5*t) + (44/3)*np.exp(-2*t) - 9*np.exp(-3*t)

plt.figure(figsize=(10,4))
plt.plot(t, vo_manual, linewidth=2)
plt.grid(True)
plt.title("Questão 5(b) - Saída v_o(t)")
plt.xlabel("t (s)")
plt.ylabel("v_o(t) [V]")
plt.show()
```

```python
t = sp.Symbol('t', real=True, positive=True)
vo = sp.Function('vo')

vi = 6*sp.exp(-3*t)

eq = sp.Eq(sp.diff(vo(t), t, 2) + 7*sp.diff(vo(t), t) + 10*vo(t),
           sp.diff(vi, t) + 6*vi)

sol = sp.dsolve(eq, vo(t), ics={vo(0): 6, sp.diff(vo(t), t).subs(t, 0): -4})
sol
```

```python
sol_simplificada = sp.simplify(sol.rhs)
sol_simplificada
```

```python
vo_analitico = (sp.Rational(1,3))*sp.exp(-5*t) + (sp.Rational(44,3))*sp.exp(-2*t) - 9*sp.exp(-3*t)

comparacao = sp.simplify(sol_simplificada - vo_analitico)
comparacao
```

```python
vo_func = sp.lambdify(t, sol_simplificada, "numpy")

tt = np.linspace(0, 2, 1000)
vo_sympy = vo_func(tt)

vo_manual = (1/3)*np.exp(-5*tt) + (44/3)*np.exp(-2*tt) - 9*np.exp(-3*tt)

plt.figure(figsize=(10,4))
plt.plot(tt, vo_sympy, label="Sympy", linewidth=2)
plt.plot(tt, vo_manual, "--", label="Analítico manual", linewidth=2)
plt.grid(True)
plt.legend()
plt.title("Questão 5(c) - Comparação Sympy vs Analítico")
plt.xlabel("t (s)")
plt.ylabel("v_o(t)")
plt.show()
```
